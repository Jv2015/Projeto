```python
#importando as bibliotecas
import pandas as pd
import numpy as np
```

```python
#importandos os daddos da base
tabela = pd.read_excel(r"C:\User\Downloads\EXP_2021_2022_20230828.xlsx")
```

Base exportada da fonte: https://docs.google.com/spreadsheets/d/1zcZ6W3AekOH8uoRnx7SLnJLT7eToBiGD/edit#gid=1322856898

```python
#verificando as colunas
print(list(tabela))
```

```python
#vendo a estrutura da tabela
tabela.head(10)
```

```python
#Tratamento dos dados referentes a Quilograms
base_column = ['Mês', 'Países', 'Bloco Econômico', 'UF do Produto', 'Via', 'URF', 'Código NCM','Descrição NCM']
second_base = ['2022 - Quilograma Líquido' , '2021 - Quilograma Líquido']
Quilograma = tabela[base_column + second_base]
Quilograma = Quilograma.melt(id_vars=base_column)

aux = Quilograma["variable"].str.split("-", 0, expand = True)
Quilograma['Ano'] = aux.get(0)
Quilograma["Dolar"] = aux.get(1)
Quilograma.drop(columns='variable', inplace= True)
Quilograma
```

```python
#Tratamento dos dados referente a Valores
base_column = ['Mês', 'Países', 'Bloco Econômico', 'UF do Produto', 'Via', 'URF', 'Código NCM','Descrição NCM']
second_base = ['2022 - Valor FOB (US$)','2021 - Valor FOB (US$)']
Valor = tabela[base_column + second_base]
Valor = Valor.melt(id_vars=base_column)

aux = Valor["variable"].str.split("-", 0, expand = True)
Valor['Ano'] = aux.get(0)
Valor["Dolar"] = aux.get(1)
Valor.drop(columns='variable', inplace= True)
Valor
```

```python
#Mesclado os dois dataframes criados, limpando as colunas que não são necessárias e renomeando as mesmas
df = pd.merge(Quilograma,Valor,how= 'left', on = ['Mês', 'Países', 'Bloco Econômico', 'UF do Produto', 'Via', 'URF', 'Código NCM','Descrição NCM', 'Ano'])
df.rename(columns={"value_x":'Quilograms', 'value_y':'Valor em US$'},inplace= True)
df.drop(columns=['Dolar_x', 'Dolar_y'], inplace= True)
df
```

```python
#Inserindo o nome dos meses no dataframe
base_mes = ['Mês']
Meses = df[base_mes].astype(int)
def mes(x):
  if x == 1:
    return "Janeiro"
  elif x == 2:
    return "Feveiro"
  elif x == 3:
    return "Março"
  elif x == 4:
    return "Abril"
  elif x == 5:
    return "Maio"
  elif x == 6:
    return "Junho"
  elif x == 7:
    return "Julho"
  elif x == 8:
    return "Agosto"
  elif x == 9:
    return "Setembro"
  elif x == 10:
    return "Outubro"
  elif x == 11:
    return "Novembro"
  elif x == 12:
    return "Dezembro"

df["Meses"] = df["Mês"].apply(mes)
df
```

```python
#Verificando como ficou
print(list(df))
```

```python

#Ordenado as colunas
colunas_ordem = ['Ano','Mês','Meses','Países', 'Bloco Econômico', 'UF do Produto', 'Via', 'URF', 'Código NCM','Descrição NCM' ,'Quilograms','Valor em US$']
df = df[colunas_ordem]
df
```

```python
base_tratada = 'base.xlsx'

df.to_excel(base_tratada, index=False)
```
